```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/processed/cleaned_telecom_data.csv')
```

```python
df['avg_rech_6_7'] = (
    df['total_rech_amt_6'] + df['total_rech_amt_7']
) / 2
```

```python
threshold = df['avg_rech_6_7'].quantile(0.7)
threshold
```

```python
high_value_df = df[df['avg_rech_6_7'] >= threshold].copy()
high_value_df.shape
```

```python
high_value_df['churn'] = np.where(
    (high_value_df['total_ic_mou_9'] == 0) &
    (high_value_df['total_og_mou_9'] == 0) &
    (high_value_df['vol_2g_mb_9'] == 0) &
    (high_value_df['vol_3g_mb_9'] == 0),
    1, 0
)
```

```python
high_value_df['churn'].value_counts(normalize=True)
```

```python
high_value_df = high_value_df.fillna(0)
```

```python
high_value_df['rech_amt_decline'] = (
    high_value_df['total_rech_amt_8'] -
    high_value_df['total_rech_amt_6']
)
```

```python
high_value_df['ic_mou_decline'] = (
    high_value_df['total_ic_mou_8'] -
    high_value_df['total_ic_mou_6']
)
```

```python
high_value_df['og_mou_decline'] = (
    high_value_df['total_og_mou_8'] -
    high_value_df['total_og_mou_6']
)
```

```python
cols_9 = [col for col in high_value_df.columns if '_9' in col]
high_value_df = high_value_df.drop(cols_9, axis=1)
```

```python
high_value_df.to_csv(
    "../data/processed/high_value_churn_data.csv",
    index=False
)
```
